```python
df.groupby('Item_Identifier')['Item_Weight'].apply(list)
```

ahi vemos que el item wieght tiene relacion con el item identifier


```python

peso_por_identifier = df.groupby('Item_Identifier')['Item_Weight'].transform('first')


df['Item_Weight'].fillna(peso_por_identifier, inplace=True)



```

en el caso de estos 4 no tenemos como saber su peso comparandolos con otros items de su mismo identificador  puesto que son unicos a lo que deberemos optar por otra opcion al ser un obnjeto en bodega no eliminaremos su existencia por lo que optaremos por darles un valor promedio segun el tipo de item


```python
peso_promedio_por_tipo = df.groupby('Item_Type')['Item_Weight'].transform('mean')
df['Item_Weight'] = df['Item_Weight'].fillna(peso_promedio_por_tipo)
```
